```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('cs_bisnode_panel.zip')
data
```

```python
data.columns
```

```python
#criteria for the holdout sample so I can mask the training sample with !holdout_criteria
holdout_criteria = (data['ind2'] == 26) & (data['year'] == 2014) & ((data['sales'] >=1000) & (data['sales']<=10000000))
```

```python
#holdout sample creation
holdout = data[holdout_criteria]
print(f'Holdout shape: {holdout.shape}')
holdout.head()
```

```python
#cleaning holdout sample
```

```python
#training sample
training = data[~holdout_criteria]
print(f'Training shape: {training.shape}')
training.head()
```

```python
#cleaning training sample:

#delete observations where sales are na (sales make up a part of the y variable)
null_sales = training['sales'].isna()
training = training[~null_sales]
training

```

```python
mask_2014 = (data.year == 2014) & (data.sales > 0)
```

```python
mask_2015 = (training.year == 2015) & ((training.sales ==0)|training.sales.isna())
```

```python
#could merge 2014 and 2015 together on comp_id. would hopefully then have 
trying = training[mask_2015]
```

```python
trying.year.value_counts()
```

```python
yo = pd.merge(holdout, trying, on = 'comp_id')
yo
```
